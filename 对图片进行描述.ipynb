```python
from sentence_transformers import SentenceTransformer
import torch
import random
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print(device)

# 载入模型，并将其转移到GPU
model = SentenceTransformer('all-mpnet-base-v2').to(device)


# 定义余弦相似度函数
def cosine_similarity(batch_embedding, single_embedding):
    return torch.nn.functional.cosine_similarity(batch_embedding, single_embedding.unsqueeze(0), dim=1)


def mpnet(data, input_sentence):
    # 生成输入句子的嵌入，并转换为tensor，然后移至GPU
    input_embedding = torch.tensor(model.encode([input_sentence])[0]).to(device)

    # 初始化结果列表
    results = []

    # 分批处理数据
    batch_size = 10  # 可以调整这个批量大小
    for i in range(0, len(data), batch_size):
        batch_data = data[i:i + batch_size]
        titles_categories = [item['title'] for item in batch_data]

        # 生成嵌入并将其移至GPU
        embeddings = model.encode(titles_categories)
        embeddings_tensor = torch.tensor(embeddings).to(device)

        # 一次计算整个批次的相似度
        similarities = cosine_similarity(embeddings_tensor, input_embedding)
        for item, similarity in zip(batch_data, similarities):
            item['confidence'] = similarity.item()
            results.append(item)

    # 根据相似度排序
    results.sort(key=lambda x: x['confidence'], reverse=True)

    # 从前25个最相似的条目中随机选择5个
    top_25 = results[:25]
    print("随机前数据", top_25)
    selected_items = random.sample(top_25, min(5, len(top_25)))

    return selected_items
```

```python
import requests
from PIL import Image
from transformers import BlipProcessor, BlipForConditionalGeneration
# 从预训练模型加载BLIP处理器
processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
# 从预训练模型加载BLIP条件生成模型
model = BlipForConditionalGeneration.from_pretrained("Salesforce/blip-image-captioning-base")

# 图片URL
img_url = 'https://migration-15tb.oss-cn-shenzhen.aliyuncs.com/tsinghua/batch6/2019-04-10/foHpRFaIOdygq6RTqeEL5A3zp.png'
# 从URL获取图片，并将其转换为RGB格式
raw_image = Image.open(requests.get(img_url, stream=True).raw).convert('RGB')

# 条件性图像描述生成
text = "a photography of"  # 描述生成的条件文本
inputs = processor(raw_image, text, return_tensors="pt")  # 处理图像和文本

# 生成描述
out = model.generate(**inputs)
# 解码并打印生成的描述
print(processor.decode(out[0], skip_special_tokens=True))


# 无条件图像描述生成
inputs = processor(raw_image, return_tensors="pt")  # 仅处理图像，无文本条件

# 生成描述
out = model.generate(**inputs)
# 解码并打印生成的描述
print(processor.decode(out[0], skip_special_tokens=True))
```

```python
from transformers import MarianMTModel, MarianTokenizer

model_name = 'Helsinki-NLP/opus-mt-en-zh'
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)

text_to_translate = "a girl in a pink dress and bunny ears"
model_inputs = tokenizer(text_to_translate, return_tensors="pt")

translated = model.generate(**model_inputs)
translated_text = tokenizer.decode(translated[0], skip_special_tokens=True)

print(translated_text)
```

```python
from PIL import Image
import numpy as np

# 由于执行环境被重置，需要重新加载图片文件
# 加载图片文件
file_path = '/content/微信图片_20231208141233.png'
image = Image.open(file_path)

# 转换图片为numpy数组
image_np = np.array(image)

# 我们将尝试找到网格的颜色并替换为白色
# 通常网格线会比背景稍暗，所以我们会找到非白色的线条
# 我们设定一个阈值来判定像素是否足够接近白色
threshold = 240

# 网格线可能不是纯黑色，所以我们找到所有不是白色的像素点
grid_mask = np.all(image_np < threshold, axis=-1)

# 我们创建一个全白图片
white_bg = np.ones_like(image_np) * 255

# 只保留不是网格线的像素
white_bg[~grid_mask] = image_np[~grid_mask]

# 将处理后的数组转换回PIL图片
white_bg_image = Image.fromarray(white_bg)

# 保存新图片
white_bg_image_path = 'white_background_no_grid.png'
white_bg_image.save(white_bg_image_path)

white_bg_image_path
```
